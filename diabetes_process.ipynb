```python
import os
import csv
import math

import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
# source: https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset
diabetes_data = pd.read_csv("./data/diabetes_og.csv")
diabetes_data.head()
```

```python
columns_to_encode = ['gender', 'age', 'smoking_history', 'bmi', 'HbA1c_level', 'blood_glucose_level']
le = preprocessing.LabelEncoder()
for column in columns_to_encode:
    diabetes_data[column] = le.fit_transform(diabetes_data[column])

diabetes_data.head()
```

```python
diabetes_data.to_csv('./data/diabetes_processed.csv', index=False)
```

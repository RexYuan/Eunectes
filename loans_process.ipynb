```python
import os
import csv
import math

import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
# source: https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset
dataset = pd.read_csv("./data/loans_og.csv", index_col=0)
dataset.head()
```

```python
columns_to_encode = ["LIMIT_BAL", "SEX", "EDUCATION", "MARRIAGE", "AGE", "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6", "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", "PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6", "default.payment.next.month"]
le = preprocessing.LabelEncoder()
for column in columns_to_encode:
    dataset[column] = le.fit_transform(dataset[column])

dataset.head()
```

```python
dataset.to_csv('./data/loans_processed.csv', index=False)
```

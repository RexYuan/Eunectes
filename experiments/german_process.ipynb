```python
import os
import csv
import math

import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
```

```python
# source: https://www.kaggle.com/datasets/kabure/german-credit-data-with-risk
dataset = pd.read_csv("./data/german_og.csv", index_col=0)
dataset.describe()
```

```python
columns_to_encode = ['Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Purpose', 'Risk']
le = preprocessing.LabelEncoder()
for column in columns_to_encode:
    dataset[column] = le.fit_transform(dataset[column])

dataset.head()
```

```python
dataset.to_csv('./data/german_processed.csv', index=False)
```
